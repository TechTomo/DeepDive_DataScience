Matplotlib Operations

```python
from matplotlib.lines import lineStyles
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

x=np.arange(0,10)
y=np.arange(11,21)


#Scatterplot -> Scatter x and y values
plt.scatter(x,y,c='r')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Graph Test')
plt.savefig('GraphTest.png')
```

```python

#Subplot

plt.subplot(2,2,1)
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.plot(x,y,'g')
plt.subplot(2,2,3)
plt.plot(x,y,'b')
```

```python

#Compute the x and y coordinates for point on a sine curve
x=np.arange(0,4*3.141,0.1)
y=np.sin(x)
plt.plot(x,y)
plt.title('sine wave')
```

```python
#Bar Plot

x1=[12,67,23]
y1=[98,85,59]

x2=[23,54,67]
y2=[45,76,12]

plt.bar(x1,y1)
plt.bar(x2,y2,color='g')

```

```python
#Histogram
a=np.array([10,24,35,47,90,53,62,6,2,7,2,52,52,61,83])
plt.hist(a)
```

```python
#Pie Chart

sizes=[200,100,200,400]
colors=['gold','green','blue','red']
labels=['C++','C',"Java",'Python']
explode=(0.1,0,0,0) #Move pie

plt.pie(sizes,explode=explode,labels=labels,colors=colors,
autopct='%1.1f%%',shadow=True)
```
